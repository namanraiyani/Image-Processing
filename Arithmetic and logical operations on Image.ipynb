```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = '/Flower_poster.png'
img = cv2.imread(image,1)
plt.axis('off')
plt.imshow(img)
```

```python
grey_img =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title('Original Image')
axes[1].imshow(grey_img, cmap = "gray")
axes[1].axis('off')
axes[1].set_title('Grey Image')
plt.show()
```

```python
x, y, w, h = 100, 100, 500, 430
crop_img = img[y : y+h, x : x+w]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title('Original Image')
axes[1].imshow(crop_img)
axes[1].axis('off')
axes[1].set_title('Cropped Image')
plt.show()
```

```python
from tensorflow.keras.datasets import mnist
(x_train, y_train), (_, _) = mnist.load_data()
```

```python
def get_random_image_for_digit(digit):
    digit_indices = np.where(y_train == digit)[0]
    digit_images = x_train[digit_indices]
    random_index = np.random.randint(0, len(digit_images))
    return digit_images[random_index]
```

```python
image_7 = 255 - get_random_image_for_digit(7)
image_3 = 255 - get_random_image_for_digit(3)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')
plt.show()
```

```python
add_result = cv2.add(image_7, image_3)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(add_result, cmap='gray')
axes[2].set_title('Add Image')
axes[2].axis('off')
plt.show()
```

```python
sub_result = cv2.subtract(image_7, image_3)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(sub_result, cmap='gray')
axes[2].set_title('Subtract Image')
axes[2].axis('off')
plt.show()
```

```python
mul_result = cv2.multiply(image_7, image_3)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(mul_result, cmap='gray')
axes[2].set_title('Multiply Image')
axes[2].axis('off')
plt.show()
```

```python
division_result = cv2.divide(image_7, image_3 + 1)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(division_result, cmap='gray')
axes[2].set_title('Divide Image')
axes[2].axis('off')
plt.show()
```

```python
alpha = 0.5
beta = 1 - alpha
blended_result = cv2.addWeighted(image_7, alpha, image_3, beta, 0)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(blended_result, cmap='gray')
axes[2].set_title('Blend Image')
axes[2].axis('off')
plt.show()
```

```python
and_result = cv2.bitwise_and(image_7, image_3)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(add_result, cmap='gray')
axes[2].set_title('Logiccal AND Image')
axes[2].axis('off')
plt.show()
```

```python
or_result = cv2.bitwise_or(image_7, image_3)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(or_result, cmap='gray')
axes[2].set_title('Logical OR Image')
axes[2].axis('off')
plt.show()
```

```python
not_result = cv2.bitwise_not(image_7)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(not_result, cmap='gray')
axes[1].set_title('Logical NOT Image')
axes[1].axis('off')
plt.show()
```

```python
xor_result = cv2.bitwise_xor(image_7, image_3)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(image_3, cmap='gray')
axes[1].set_title('Digit 3')
axes[1].axis('off')

axes[2].imshow(xor_result, cmap='gray')
axes[2].set_title('Logical XOR Image')
axes[2].axis('off')
plt.show()
```

```python
leftshift_result = np.left_shift(image_7, 6)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_7, cmap='gray')
axes[0].set_title('Digit 7')
axes[0].axis('off')

axes[1].imshow(leftshift_result, cmap='gray')
axes[1].set_title('Bitwiise Left Shift Image')
axes[1].axis('off')
plt.show()
```

```python
rightshift_result = np.right_shift(image_3, 7)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_3, cmap='gray')
axes[0].set_title('Digit 3')
axes[0].axis('off')

axes[1].imshow(rightshift_result, cmap='gray')
axes[1].set_title('Bitwiise Right Shift Image')
axes[1].axis('off')
plt.show()
```

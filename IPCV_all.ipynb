```python
def add_images(img_array1, img_array2):
    rows, cols = img_array1.shape
    added_image = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            added_image[i, j]= min(img_array1[i,j] + img_array2[i,j], 255)
    return added_image
display_3_images(img_array1, img_array2, add_images(img_array1, img_array2))
```

```python
def sub_images(img_array1, img_array2):
    rows, cols = img_array1.shape
    sub_img = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            sub_img[i, j] = max(0, img_array1[i,j] - img_array2[i,j])
    return sub_img
display_3_images(img_array1, img_array2, sub_images(img_array1, img_array2))
```

```python
def mul_images(img_array1, img_array2):
    rows, cols = img_array1.shape
    mul_image = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            mul_image[i, j]= min(img_array1[i,j] * img_array2[i,j], 255)
    return mul_image
display_3_images(img_array1, img_array2, mul_images(img_array1, img_array2))
```

```python
def div_images(img_array1, img_array2):
    rows, cols = img_array1.shape
    div_image = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            div_image[i, j]= min(img_array1[i,j] // (img_array2[i,j] + 1), 255)
    return div_image
display_3_images(img_array1, img_array2, div_images(img_array1, img_array2))
```

```python
def blend_image(img_array1, img_array2, alpha = 0.5):
    rows, cols = img_array1.shape
    blended = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            blended[i,j] = int(alpha*img_array1[i,j] + (1-alpha)*img_array2[i,j])
    return blended
display_3_images(img_array1, img_array2, blend_image(img_array1, img_array2))
```

```python
def bitwise_and(img_array1, img_array2):
    rows, cols = img_array1.shape
    img_array1 = img_array1.astype(np.uint8)
    img_array2 = img_array2.astype(np.uint8)
    and_img = np.zeros((rows,cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            and_img[i, j] = img_array1[i,j] & img_array2[i,j]
    return and_img
display_3_images(img_array1, img_array2, bitwise_and(img_array1, img_array2))
```

```python
def bitwise_or(img_array1, img_array2):
    rows, cols = img_array1.shape
    img_array1 = img_array1.astype(np.uint8)
    img_array2 = img_array2.astype(np.uint8)
    or_img = np.zeros((rows,cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            or_img[i, j] = img_array1[i,j] | img_array2[i,j]
    return or_img
display_3_images(img_array1, img_array2, bitwise_or(img_array1, img_array2))
```

```python
def bitwise_xor(img_array1, img_array2):
    rows, cols = img_array1.shape
    img_array1 = img_array1.astype(np.uint8)
    img_array2 = img_array2.astype(np.uint8)
    xor_img = np.zeros((rows,cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            xor_img[i, j] = img_array1[i,j] ^ img_array2[i,j]
    return xor_img
display_3_images(img_array1, img_array2, bitwise_xor(img_array1, img_array2))
```

```python
def digit_negative(img):
    rows, cols = img.shape
    img = img.astype(np.uint8)
    neg = np.zeros((rows,cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            neg[i,j] = 255-img[i,j]
    return neg
display_2_images(img_array1,digit_negative(img_array1))
```

```python
def thresholding(img, thresholding = 127):
    rows,cols=img.shape
    new = np.zeros((rows,cols),dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            if img[i,j] >= thresholding:
                new[i,j]=255
    return new
display_2_images(img_array1, thresholding(img_array1))
```

```python
def global_thresholding_without_bg(img,a,b):
    rows,cols = img.shape
    img = img.astype(np.uint8)
    new = np.zeros((rows,cols),dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            if a<=img[i,j]<=b:
                new[i,j] = 255
    return new
display_2_images(img_array1, global_thresholding_without_bg(img_array1, 200,255))
```

```python
def global_thresholding_with_bg(img,a,b):
    rows,cols = img.shape
    img = img.astype(np.uint8)
    new = np.zeros((rows,cols),dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            if a<=img[i,j]<=b:
                new[i,j] = img[i,j]
    return new
display_2_images(img_array1, global_thresholding_with_bg(img_array1, 150,255))
```

```python
def contrast_stretching(img, r1,s1,r2,s2):
    img = img.astype(np.uint8)
    new = ((img-r1)*(s2-s1) / (r2-r1)) + s1
    new = np.clip(new,0,255)
    return new
display_2_images(img_array1,contrast_stretching(img_array1, 10,20,30,40))
```

```python
def log_transformation(img, c=30):
    img = img.astype(np.uint8)
    new = np.log1p(img)*c
    new = np.clip(new,0,255)
    return new
display_2_images(img_array1, log_transformation(img_array1))
```

```python
def normalize(img):
    minimum, maximum = np.min(img), np.max(img)
    normalized = ((img-minimum) / (maximum-minimum))*255
    return normalized
def gamma_correction(img, c=2.0, gamma=1.5):
    img = normalize(img)
    gamma_corrected = c * (img / 255.0) ** gamma
    gamma_corrected = np.clip(gamma_corrected * 255 , 0 , 255)
    return gamma_corrected
display_2_images(img_array1, gamma_correction(img_array1))
```

```python
def display_and_plot_histogram(img1):
    fig, ax = plt.subplots(1, 2, figsize=(6,3))
    ax[0].imshow(img1, cmap='gray')
    ax[0].axis('off')
    ax[0].set_title('original_image')
    ax[1].hist(img1.ravel() * 255, bins=256, range=(0, 256))
    ax[1].set_xlabel('pixel_value')
    ax[1].set_ylabel('frequency')
```

```python
def get_pixel_freq(img):
    pix_freq = np.zeros(256)
    rows, cols = img.shape
    for i in range(rows):
        for j in range(cols):
            pixel = int(img[i,j])
            pix_freq[pixel] += 1
    return pix_freq

def histogram_equalization(img):
    L = 256
    pixel_freq = get_pixel_freq(img)
    pixel_pdf = pixel_freq / np.sum(pixel_freq)
    pixel_cumsum = np.cumsum(pixel_pdf)
    bits_cumsum = (L-1)*pixel_cumsum
    pixel_map = np.fix(bits_cumsum).astype(int)
    rows, cols = img.shape
    output_image=np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            pixel_val = int(img[i,j])
            output_image[i,j] = pixel_map[pixel_val] / 255
    return output_image

equalized_image = histogram_equalization(img_array1)
display_and_plot_histogram(img_array1)
```

```python
def apply_histogram_stretching(img):
    img = img.astype(np.float32)
    rmin, rmax = np.min(img), np.max(img)
    smin, smax = 0, 255
    img_stretched = ((smax-smin) / (rmax-rmin)) * (img-rmin) + smin
    img_stretched.astype(np.uint8)
    img_stretched = np.clip(img_stretched,0,255)
    return img_stretched

stretched_image = apply_histogram_stretching(img_array1)
display_and_plot_histogram(stretched_image)
```

```python
def add_gaussian_noise(img, mean=0, std_dev=75):
    rows,cols=img.shape
    gaussian_noise = np.random.normal(mean,std_dev,(rows,cols))
    noisy_img = np.clip(img + gaussian_noise, 0, 255)
    noisy_img = noisy_img.astype(np.uint8)
    return noisy_img

def add_sal_and_peper_noise(img):
    rows,cols=img.shape
    noisy_img=np.copy(img)
    salt_pixels = np.random.randint(15,40)
    for _ in range(salt_pixels):
        y = np.random.randint(0, rows-1)
        x = np.random.randint(0, cols-1)
        noisy_img[y,x]=255
    pepper_pixels = np.random.randint(15,40)
    for _ in range(pepper_pixels):
        y = np.random.randint(0, rows-1)
        x = np.random.randint(0, cols-1)
        noisy_img[y,x] = 0
    return noisy_img
gauss_img = add_gaussian_noise(img_array1)    
sal_n_pepper = add_sal_and_peper_noise(img_array1)
display_3_images(img_array1, gauss_img, sal_n_pepper)
```

```python
def avg_filtered_image(img):
    mask_size = 3
    mask = np.ones((mask_size, mask_size)) / (mask_size ** 2)
    rows,cols = img.shape
    mask_height, mask_width = mask.shape
    pad_height, pad_width = mask_height // 2, mask_width // 2
    padded_img = np.pad(img, 
                        ((pad_height, pad_height),
                        (pad_width, pad_width)
                       ),
                        mode = 'constant',
                        constant_values = 0
                        )
    filtered_img = np.zeros((rows,cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            region = padded_img[i:i+mask_height, j:j+mask_width]
            filtered_img[i,j] = np.sum(region * mask)
    return filtered_img

def median_filter(img):
    mask_size = 3
    mask = np.ones((mask_size, mask_size)) / (mask_size ** 2)
    rows, cols = img.shape
    mask_height, mask_width = mask.shape
    pad_height, pad_width = mask_height // 2, mask_width // 2
    padded_img = np.pad(img,
                       ((pad_height, pad_height),
                        (pad_width, pad_width)
                       ),
                        mode='constant',
                        constant_values=0
                       )
    filtered_img = np.zeros((rows, cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            region = padded_img[i:i+mask_height, j:j+mask_width]
            filtered_img[i,j] = np.median(region)
    return filtered_img

avg_filtered = avg_filtered_image(gauss_img)
med_filtered = median_filter(sal_n_pepper)
display_3_images(img_array1, avg_filtered, med_filtered)
```

```python
def high_pass(img, kernel):
    kernel_size = 3
    padding = kernel_size // 2
    padded_img = np.pad(
        img,
        ((padding,padding),
        (padding,padding)),
        mode = 'constant',
        constant_values=0
    )
    rows, cols = img.shape
    filtered_img = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            region = padded_img[i:i+kernel_size, j:j+kernel_size]
            filtered_img[i,j] = np.sum(region * kernel)
    filtered_img = filtered_img.astype(np.uint8)
    filtered_img = np.clip(filtered_img, 0, 255)
    return filtered_img

def high_boost(img, kernel, A):
    high_pass_img = high_pass(img, kernel)
    high_pass_img = (A-1) * img + high_pass_img
    high_pass_img = np.clip(high_pass_img, 0, 255).astype(np.uint8)
    return high_pass_img
    
kernel = [
    [-1, -1, -1],
    [-1, 9, -1],
    [-1, -1, -1],
]
high_pass_img = high_pass(img_array1, kernel)
high_boost_img = high_boost(img_array1, kernel, A=2.1)               
display_3_images(img_array1, high_pass_img, high_boost_img)
```

```python
def erosion(img, se):
    rows, cols = img.shape
    se_h, se_w = se.shape
    se_center_y, se_center_x = se_h // 2, se_w //2
    eroded_img = np.zeros((rows,cols),dtype = np.uint8)
    for y in range(rows):
        for x in range(cols):
            eroded_img[y,x] = 255
            local_min = 255
            for j in range(-se_center_y, se_center_y+1):
                for i in range(-se_center_x, se_center_x+1):
                    if se[j + se_center_y, i + se_center_x] == 1:
                        img_y = y + j
                        img_x = x + i
                        if 0 <= img_y < rows and 0 <= img_x < cols:
                            local_min = min(local_min, img[img_y, img_x])
            eroded_img[y,x] = local_min
    return eroded_img

def dilation(img, se):
    rows, cols = img.shape
    se_h, se_w = se.shape
    se_center_y, se_center_x = se_h // 2, se_w  //2
    dilated = np.zeros((rows, cols), dtype = np.uint8)    
    for y in range(rows):
        for x in range(cols):
            local_max = 0
            for j in range(-se_center_y, se_center_y+1):
                for i in range(-se_center_x, se_center_x+1):
                    if se[j+se_center_x, se_center_x+1]:
                        img_y = y+j
                        img_x = x+i
                        if 0<=img_y<rows and 0<=img_x<cols:
                            local_max = max(local_max, img[img_y, img_x])
            dilated[y,x] = local_max
    return dilated

se = np.array([
    [1,1,1],
    [1,1,1],
    [1,1,1],
])
display_3_images(img_array1, erosion(img_array1,se), dilation(img_array1, se))
```

```python
def closing(img, se):
    return erosion(dilation(img, se),se)
def opening(img, se):
    return dilation(erosion(img, se),se)
display_3_images(img_array1, closing(img_array1, se), opening(img_array1,se))
```

```python
def hit_and_miss(img, se, background):
    rows, cols = img.shape
    se_height, se_width = se.shape
    se_center_y, se_center_x = se_height // 2, se_width // 2
    res = np.zeros((rows, cols), dtype = np.uint8)
    for y in range(rows):
        for x in range(cols):
            match = True
            for j in range(-se_center_y, se_center_y+1):
                for i in range(-se_center_x, se_center_x+1):
                    img_y = y+j
                    img_x = x+i
                    if se[j + se_center_y, i+se_center_x] == 1:    
                        if 0<=img_y<rows and 0<=img_x<cols:
                            if img[img_y,img_x] == 0:
                                match = False
                    elif se[j+se_center_y, i + se_center_x] == 0:
                        if 0<=img_y<rows and 0<=img_x<cols:
                            if img[img_y, img_x] != background:
                                match = False
                if match:
                    res[y,x] = 255
    return res
se_neg = np.where(se == 1, 0, 1)
display_2_images(img_array1, hit_and_miss(img_array1, se_neg, 255))
```

```python
from scipy.fft import fft2, ifft2
def compute_fourier_transform(img):
    rows, cols = img.shape
    x_coords, y_coords = np.meshgrid(np.arange(cols), np.arange(rows))
    transformed_image = img*(-1)**(x_coords+y_coords)
    return fft2(transformed_image)
fourier_image = compute_fourier_transform(img_array1)
display_2_images(img_array1, 20*np.log(1+abs(fourier_image)))
```

```python
def low_pass_filter(fourier_transform, cutoff):
    rows,cols = fourier_transform.shape
    u_coords, v_coords = np.meshgrid(np.arange(cols) - cols // 2, np.arange(rows) - rows // 2)
    distance_from_origin = np.sqrt(u_coords**2 + v_coords**2)
    low_pass_mask = (distance_from_origin <= cutoff)
    filtered_image = fourier_transform * low_pass_mask
    output = ifft2(filtered_image).real
    output = output*(-1)**(np.arange(rows)[:,None] + np.arange(cols))
    return output, low_pass_mask
cutoff_freq=10
filtered_image, lp_mask = low_pass_filter(fourier_image,cutoff_freq)
display_3_images(img_array1, filtered_image, lp_mask)
```

```python
def high_pass_filter(fourier_transform, cutoff):
    rows,cols = fourier_transform.shape
    u_coords, v_coords = np.meshgrid(np.arange(cols) - cols // 2, np.arange(rows) - rows // 2)
    distance_from_origin = np.sqrt(u_coords**2 + v_coords**2)
    low_pass_mask = (distance_from_origin > cutoff)
    filtered_image = fourier_transform * low_pass_mask
    output = ifft2(filtered_image).real
    output = output*(-1)**(np.arange(rows)[:,None] + np.arange(cols))
    return output, low_pass_mask
cutoff_freq=10
filtered_image, lp_mask = high_pass_filter(fourier_image,cutoff_freq)
display_3_images(img_array1, filtered_image, lp_mask)
```

```python
def gaussian_low_pass_filter(img, fourier_transform, cutoff):
    rows,cols = fourier_transform.shape
    u_coords, v_coords = np.meshgrid(np.arange(cols) - cols // 2, np.arange(rows) - rows // 2)
    distance_from_origin = np.sqrt(u_coords**2 + v_coords**2)
    mask = np.exp(-(distance_from_origin**2) / (2*(cutoff**2)))
    output = ifft2(mask*fourier_transform).real
    output = output*(-1)**(rows+cols)
    return output, mask
gaussian_lp_image, gaussian_lp_mask = gaussian_low_pass_filter(img_array1, fourier_image, cutoff=60)
display_3_images(img_array1, gaussian_lp_image, gaussian_lp_mask)
```

```python
def gaussian_high_pass_filter(img, fourier_transform, cutoff):
    rows,cols = fourier_transform.shape
    u_coords, v_coords = np.meshgrid(np.arange(cols) - cols // 2, np.arange(rows) - rows // 2)
    distance_from_origin = np.sqrt(u_coords**2 + v_coords**2)
    mask = 1-np.exp(-(distance_from_origin**2) / (2*(cutoff**2)))
    output = ifft2(mask*fourier_transform).real
    output = output*(-1)**(rows+cols)
    return output, mask
gaussian_hp_image, gaussian_hp_mask = gaussian_high_pass_filter(img_array1, fourier_image, cutoff=60)
display_3_images(img_array1, gaussian_hp_image, gaussian_hp_mask)
```

```python
from collections import deque
def region_growing(img, seed, threshold):
    rows, cols=img.shape
    binary_image = np.zeros((rows,cols), dtype=np.uint8)
    x,y=seed
    mean_intensity = img[x,y]
    region = deque([(x,y)])
    visited = set()
    directions = [(-1,0), (1,0), (0,-1), (0,1)]
    while region:
        cx,cy=region.popleft()
        if (cx,cy) in visited:
            continue
        visited.add((cx,cy))
        binary_image[cx,cy] = 255
        for dx, dy in directions:
            nx,ny=cx+dx, cy+dy
            if 0<=nx<rows and 0<=ny<cols:
                if (nx,ny) not in visited and abs(int(img[nx,ny]) - mean_intensity) <= threshold:
                    region.append((nx, ny))
    return binary_image
seed = (13, 13)
threshold = 200
display_2_images(img_array1, region_growing(img_array1, seed, threshold))
```

```python
def region_split_merge(img, intensity_threshold):
    def recursive_divide_merge(sub):
        h, w = sub.shape
        if h<=2 or w<=2:
            return sub
        if np.max(sub) - np.min(sub) <= intensity_threshold:
            return np.full(sub.shape, 128, dtype=np.uint8)
        mid_h, mid_w = h//2, w//2
        top_left_part = recursive_divide_merge(sub[:mid_h, :mid_w])
        top_right_part = recursive_divide_merge(sub[:mid_h, mid_w:])
        bottom_left_part = recursive_divide_merge(sub[mid_h:, :mid_w])
        bottom_right_part = recursive_divide_merge(sub[mid_h:, mid_w:])
        return np.vstack(
            (np.hstack((top_left_part, top_right_part)),
            np.hstack((bottom_left_part, bottom_right_part))),
        )
    return recursive_divide_merge(img)
threshold = 25
merged_img = region_split_merge(img_array1, threshold)
display_2_images(img_array1, merged_img)
```
